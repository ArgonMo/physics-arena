# Random walk (1d, 2d and 3d Random walk)
1. First step is importing libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
```

2. 1d random walk function:

```python
def random1d(n):
    x = 0
    d = []

    for i in range(n+1):
        ch = np.random.rand()
        if ch > 0.5:
            x += 1
        else:
            x -= 1
        d.append(x)
    plt.plot(d)
    plt.show()
```

3. simulate 1d randomwalk:

```python
n=int(input("Enter number of steps:"))
random1d(n)
```

4. 2d random walk function:

```python
def random2d(n):
    x, y = 0, 0
    dx = []
    dy = []

    choices = ['Up', 'Down', 'Left', 'Right']

    for i in range(n+1):
        ch = np.random.choice(choices)

        if ch == 'Up':
            y += 1
        elif ch == 'Down':
            y -= 1
        elif ch == 'Right':
            x += 1
        elif ch == 'Left':
            x -= 1

        dx.append(x)
        dy.append(y)
    print(f'final x is :{x} \n final y is :{y}')
    plt.plot(dx, dy)
    plt.show()
```

5. simulate 2d randomwalk:

```python
n=int(input("Enter number of steps:"))
random2d(n)
```

6. 3d random walk function:

```python
def random3d(n):
    x, y, z = 0, 0, 0
    dx = []
    dy = []
    dz = []

    choices = ['Up', 'Down', 'Left', 'Right', 'Front', 'Back']

    for i in range(n+1):
        ch = np.random.choice(choices)
        dx.append(x)
        dy.append(y)
        dz.append(z)

        if ch == 'Up':
            z += 1
        elif ch == 'Down':
            z -= 1
        elif ch == 'Right':
            x += 1
        elif ch == 'Left':
            x -= 1
        elif ch == 'Back':
            y -= 1
        elif ch == 'Front':
            y += 1

    R = np.sqrt((x**2) + (y**2) + (z**2))  # Distance from first point

    print(f'Final point:\nx :{x} \ny :{y}\nz :{z}')  # Final point

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    plt.plot(dx, dy, dz)
    plt.show()
```

7. simulate 3d randomwalk:

```python
n=int(input("Enter number of steps:"))
random3d(n)
```
